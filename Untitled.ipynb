```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import cv2
```

```python
seis = cv2.imread('00_Seismic.png')
seis = cv2.cvtColor(seis,cv2.COLOR_BGR2RGB)
```

```python
seis
```

```python
seis.shape
```

```python
saltdome = cv2.imread('00_Saltdome.png')
plt.imshow(saltdome)
```

```python
subparallel = cv2.imread('00_Subparallel.png')
plt.imshow(subparallel)
```

```python
anticline = cv2.imread('01_Anticline.png')
plt.imshow(anticline)
```

```python
methods = ['cv2.TM_CCOEFF','cv2.TM_CCOEFF_NORMED','cv2.TM_CCORR','cv2.TM_CCORR_NORMED','cv2.TM_SQDIFF','cv2.TM_SQDIFF_NORMED']
```

```python
for m in methods:
    
    # CREATE A COPY 
    seis_copy = seis.copy()
    
    method = eval(m)
    
    # TEMPLATE MATCHING
    res = cv2.matchTemplate(seis_copy,anticline,method)
    
    min_val,max_val,min_loc,max_loc = cv2.minMaxLoc(res)
    
    if method in [cv2.TM_SQDIFF,cv2.TM_SQDIFF_NORMED]:
        top_left = min_loc
    else: 
        top_left = max_loc
        
    height,width,channels = anticline.shape
    
    bottom_right = (top_left[0]+width,top_left[1]+height)
    
    cv2.rectangle(seis_copy,top_left,bottom_right,(255,0,0),10)
    
    # PLOT AND SHOW THE IMAGES
    
    plt.subplot(121)
    plt.imshow(res)
    plt.title('HEATMAP OF TEMPLATE MATCHING')
    
    plt.subplot(122)
    plt.imshow(seis_copy)
    plt.title('DETECTION OF TEMPLATE')
    
    # TEMPLATE WITH THE METHOD USED 
    plt.suptitle(m)
    
    plt.show()
    
    print('\n')
    print('\n')
```
